```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from statsmodels.tsa.holtwinters import ExponentialSmoothing
sns.set()
```

```python
airline = pd.read_csv('Airlines.csv')
airline
```

```python
airline.info()
```

```python
data = airline.copy()
data['t'] = range(1, len(data)+1)
data['t_square'] = data['t']**2
data['Date'] = pd.to_datetime(data['Month'], format="%b-%y")
data['month'] = data['Date'].dt.strftime('%b')
data['year'] = data['Date'].dt.strftime('%Y')
data
```

```python
plt.figure(figsize = (10,10))
plt.plot(data['Passengers'])
plt.show()
```

```python
sns.lineplot(x = 'year', y = 'Passengers', data = data)
```

```python
data_1 = data.drop(columns = ['Month', 'Date'])
data_dummies = pd.get_dummies(data_1)
data_dummies
```

```python
#QUADRATIC MODEL
model = LinearRegression()
model.fit(data_dummies[['t', 't_square']][0:84], data_dummies['Passengers'][0:84])
pred = model.predict(data_dummies[['t', 't_square']][84:])
rmse = np.sqrt(np.mean((data_dummies['Passengers'][84:] - pred)**2))
rmse
```

```python
#Added seasonality and quadratic log model
X = data_dummies.drop('Passengers', axis = 1)
model = LinearRegression()
model.fit(X[0:84], np.log(data_dummies['Passengers'][0:84]))
pred = model.predict(X[84:])
rmse = np.sqrt(np.mean((data_dummies['Passengers'][84:] - np.exp(pred))**2))
rmse
```

```python
model = LinearRegression()
model.fit(X[0:84], data_dummies['Passengers'][0:84])
pred = model.predict(X[84:])
rmse = np.sqrt(np.mean((data_dummies['Passengers'][84:] - pred)**2))
rmse
```

```python
pred_df = pd.DataFrame()
pred_df['Month'] = data['Month'][84:]
pred_df['Actual'] = data_dummies['Passengers'][84:]
pred_df['Predicted'] = pred
pred_df['% Error'] = abs(data_dummies['Passengers'][84:] - pred)/data_dummies['Passengers'][84:]*100
pred_df
```
